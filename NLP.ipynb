```python
data=df['time']
```

```python
data1=data.tolist()
```

```python
def preparedata(data,steps):
  x=[]
  y=[]
  for i in range(len(data)-steps):
    x.append(data[i:i+3])
    y.append(data[i+3])
  return x,y
```

```python
x,y=preparedata(data,3)
```

```python
import numpy as np
xarr=np.array(x)
yarr=np.array(y)
yarr1=yarr.reshape(-1,1)
```

```python
from keras.layers import Flatten,Dense,LSTM
from keras.models import Sequential
model=Sequential()
model.add(LSTM(200,activation='relu',return_sequences=True,input_shape=(3,1)))
model.add(LSTM(200))
model.add(Dense(42,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(xarr,yarr1,epochs=40)
```

```python
pred=model.predict(xarr)
```
